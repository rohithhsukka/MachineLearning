```python
df1 = df1[df1['D'] != 'l']
df1 = df1[df1['E'] != 'gg']
df1 = df1[df1['F'] != 'r']
df1 = df1[df1['G'] != 'o']

df1.nunique()
```

```python
#Label encoding for columns with only two category

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in ['A','D','E','I','J','L']:
  df1[col] = le.fit_transform(df1[col])
```

```python
df1.head()
```

```python
df1.dtypes
```

```python
#One hot encoding for 'F','G','M'

df1 = pd.get_dummies(df1 , columns=['F','G','M'])
```

```python
df1.head()
```

```python
df1['Approval'] = df1['Approval'].map({'+':1 , '-':0})
```

```python
df1.dtypes
```

```python
#df1.select_dtypes(include=['object'])
```

```python
X = df1.drop('Approval' , axis=1)
Y = df1['Approval']
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , Y_train , Y_test = train_test_split(X , Y, test_size=0.2 , random_state=42)
```

```python
#Scale using min-max-scaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
model1 = LogisticRegression()
model2 = KNeighborsClassifier(n_neighbors=7)
model3 = DecisionTreeClassifier(max_depth=3)
model4 = RandomForestClassifier(max_depth=3)
```

```python
model1.fit(X_train , Y_train)
model2.fit(X_train , Y_train)
model3.fit(X_train , Y_train)
model4.fit(X_train , Y_train)
```

```python
#Training data accuracy for all models
print(model1.score(X_train , Y_train))
print(model2.score(X_train , Y_train))
print(model3.score(X_train , Y_train))
print(model4.score(X_train , Y_train))
```

```python
#Test Data accuracy
print(model1.score(X_test , Y_test))
print(model2.score(X_test , Y_test))
print(model3.score(X_test , Y_test))
print(model4.score(X_test , Y_test))
```

```python
#Confusion metrics for model

from sklearn.metrics import confusion_matrix

print("Logistic-Regression - ")
print(confusion_matrix(Y_test , model1.predict(X_test)))
print()

print("KNeighborsClassifier - ")
print(confusion_matrix(Y_test , model2.predict(X_test)))
print()

print("DecisionTreeClassifier -")
print(confusion_matrix(Y_test , model3.predict(X_test)))
print()

print("RandomForestClassifier - ")
print(confusion_matrix(Y_test , model4.predict(X_test)))
print()
```

```python
#roc curve analysis

from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

fprlog,tprlog,_ = roc_curve(Y_test , model1.predict_proba(X_test)[:,1])
fprknn,tprknn,_ = roc_curve(Y_test , model2.predict_proba(X_test)[:,1])
fprdt,tprdt,_ = roc_curve(Y_test , model3.predict_proba(X_test)[:,1])
fprrf,tprrf,_ = roc_curve(Y_test , model4.predict_proba(X_test)[:,1])
```

```python
auc_log = roc_auc_score(Y_test , model1.predict_proba(X_test)[:,1])
auc_knn = roc_auc_score(Y_test , model2.predict_proba(X_test)[:,1])
auc_dt = roc_auc_score(Y_test , model3.predict_proba(X_test)[:,1])
auc_rf = roc_auc_score(Y_test , model4.predict_proba(X_test)[:,1])
```

```python
#plot the curve

plt.figure(figsize=(12,6))

plt.plot(fprlog , tprlog , label=f'Logistic Regression - {auc_log}' , marker = '*')
plt.plot(fprknn , tprknn , label=f'KNeighborsClassifier - {auc_knn}' , marker = '*')
plt.plot(fprdt , tprdt , label=f'DecisionTreeClassifier - {auc_dt}' , marker = '*')
plt.plot(fprrf , tprrf , label=f'RandomForestClassifier - {auc_rf}' , marker = '*')

plt.legend()
plt.show()
```
