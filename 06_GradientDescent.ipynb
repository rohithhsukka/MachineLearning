**Gradient Descent**
  * An error optimization process in machine learning

```python
import numpy as np

x = np.array([4 , 7 , 5.5 , 6 , 8 , 11])          #adv expenses
y  = np.array([4.5 , 8 , 5.2 , 7 , 11 , 13])      #sales revenue 
```

```python
import matplotlib.pyplot as plt

plt.scatter(x,y)

plt.show()
```

```python
#equation of linear line => yh = w0 + w1 * xabs

w0 = 2
w1 = 2

yh = w0 + w1 * x
print(yh)
```

```python
plt.scatter(x,y)

plt.plot(x,yh,'r')

plt.show()
```

```python
error = ((y - yh)**2).mean()
print(error)
```

```python
#Calculate Gradients

dew0 = -2 * ((y-yh)).mean()
dew1 = -2 * ((y-yh) * x).mean()

print(dew0)
print(dew1)
```

```python
#change the weights value using gradients
lr = 0.001
w0 = w0 - lr * dew0
w1 = w1 - lr * dew1

yh = w0 + w1 * x
error = ((y - yh)**2).mean()
print(error)
```

```python
w0 = 0.1 
w1 = 0.1

for i in range(0,50):
    yh = w0 + w1 * x

    dew0 = -2 * (y-yh).mean()
    dew1 = -2 * ((y - yh) * x).mean()

    lr = 0.001
    w0 = w0 - lr * dew0
    w1 = w1 - lr * dew1

    error = ((y - yh)**2).mean()
    print(i , error)
```

**Implement Gradient Descent on Advertising Data**

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/bipulshahi/Dataset/main/Advertising.csv',
                 index_col = 0)

df.head()
```

```python
X = df[['TV','radio']]
Y = df['sales']
```

```python
w0 = 2
w1 = 3
w2 = 4

Yh = w0 + w1 * X.TV + w2 * X.radio
print(Yh)
```

```python
error = ((Y - Yh)**2).mean()
print(error)
```

```python
#apply gradient descent to tune the coefficients and optimize the error

dew0 = -2 * ((Y - Yh)).mean()
dew1 = -2 * ((Y - Yh) * X.TV).mean()
dew2 = -2 * ((Y - Yh) * X.radio).mean()

print(dew0 , dew1 , dew2)
```

```python
lr = 0.00001
w0 = w0 - lr * dew0
w1 = w1 - lr * dew1
w2 = w2 - lr * dew2

Yh = w0 + w1 * X.TV + w2 * X.radio

error = ((Y - Yh)**2).mean()
print(error)
```

```python
w0 = 2
w1 = 3
w2 = 4

for i in range(0,1000):
  Yh = w0 + w1 * X.TV + w2 * X.radio

  dew0 = -2 * ((Y - Yh)).mean()
  dew1 = -2 * ((Y - Yh) * X.TV).mean()
  dew2 = -2 * ((Y - Yh) * X.radio).mean()

  lr = 0.00001
  w0 = w0 - lr * dew0
  w1 = w1 - lr * dew1
  w2 = w2 - lr * dew2

  Yh = w0 + w1 * X.TV + w2 * X.radio

  error = ((Y - Yh)**2).mean()
  print(i , error)
```

```python
print(w0 , w1 , w2)
```
