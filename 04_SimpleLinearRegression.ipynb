**Supervised Machine Learning - Simple Linear Regression**

**Regression - Linear**

```python
import numpy as np

x = np.array([3, 6, 4.5, 7, 9, 12])      #adv expenses
y = np.array([4.5, 8, 5.2, 7, 11, 13])      #sales revenue
```

```python
import matplotlib.pyplot as plt

plt.scatter(x, y)

plt.show()
```

**Least Square Method**

```python
#Slope of best fit line with minimum distance from all given parts

n1 = ((x - x.mean()) * (y - y.mean())).sum()
d1 = ((x - x.mean())**2).sum()

m = n1/d1
print(m)

#intercept of best fit line
c = y.mean() - m * x.mean()
print(c)
```

```python
#equation of best fit line
yh = m * x +c
print(yh)
```

```python
plt.scatter(x, y)

plt.plot(x , yh , color = 'r')

plt.show()
```

```python
#predict the value of y(sales) if value of x(adv cost)=8.9
sales = m * 8.9 + c
print(sales)
```

```python
#calculate average error => average of difference between real value of y and calculated value of y
#real sales value in the data - predicted sales value using calculated coefficient & intercept
#mean absolute Error 


mae = (abs(y - yh)).mean()
print(mae)
```

**Linear Regression to predict salary of employee**

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/rohithsukka/DataSets/main/Salary_dataset.csv', index_col = 0)
df.head()
```

* using simple linear regression to find out coefficient and intercepts to preditct employee salary

```python
x = df['YearsExperience']
y = df['Salary']
```

```python
import matplotlib.pyplot as plt

plt.scatter(x, y)

plt.show()
```

```python
#Ordinary Least Square for Simple Linear Regression


n1 = ((x - x.mean()) * (y - y.mean())).sum()
d1 = ((x - x.mean())**2).sum()

m = n1/d1
print(m)

#intercept of best fit line
c = y.mean() - m * x.mean()
print(c)
```

```python
#equation of best fit line
yh = m * x +c
print(yh)
```

```python
yh = m * x + c

plt.scatter(x,y)

plt.plot(x,yh,color='r')

```

```python
#real salary - calculated salary for each data point

mae = (abs(y - yh)).mean()
print(mae)
```

```python
print(m, " ", c)
```

```python
#predict the salary of a new hiring if year of experience is 9.2

yoe = 9.2
predicted_salary = m * 9.2 + c
print(predicted_salary)
```

**Implement Simple Linear Regression using SciKit-Learn**

```python
df.head()
```

```python
X = df[['YearsExperience']]
Y = df[['Salary']]

#split the data i.e X and Y into train and test 
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,train_size=0.75)

#import library to train 
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()

#Fit x and y to train 
model1.fit(xtrain,ytrain)

#print the training coefficients and intercepts 
print("coef or slope" , model1.coef_)
print("intercept" , model1.intercept_)
```

```python
#check model performance on train data and test data 

#prediction on data used for training or data used for training
ytrainPred = model1.predict(xtrain)

##prediction on data not used for training or test data 
ytestPred = model1.predict(xtest)

#differnce between real salary and predicted salary
mae_trainig = abs(ytrain - ytrainPred).mean()
mae_test = abs(ytest - ytestPred).mean()

print("Mean absolute error(Train)",mae_trainig)
print("Mean absolute error(Test)",mae_test)
```

```python
#predict the salary for yoe = 9.2

print(model1.predict([[9.2]]))
```
